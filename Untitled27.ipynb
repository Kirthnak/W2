```python
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
```

```python
scaler.fit(df)
```

```python
# Transform the DataFrame using the fitted scaler
scaled_data = scaler.transform(df)

# Convert the scaled data back to a DataFrame
scaled_df = pd.DataFrame(scaled_data, columns=df.columns)

print(scaled_df)
```

```python
scaled_df.drop(columns='lights', axis=1)
```

```python
scaled_df = scaled_df.round(2)
```

```python
y=scaled_df['Appliances']
x=scaled_df.drop(columns = ['Appliances'], axis=1)
```

```python
#splitting the data into 80/20 ration
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.20, random_state=3)
```

```python
#Define the model
linear_model = LinearRegression()

#fit our model
linear_model.fit(x_train, y_train)

#prediction
y_pred = linear_model.predict(x_test)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
rmse = mean_squared_error(y_test, y_pred, squared=False)  # Calculate RMSE from MSE
r2 = r2_score(y_test, y_pred)

# Print the metrics
print("Mean Absolute Error:", mae)
print("Mean Squared Error:", mse)
print("Root Mean Squared Error:", rmse)
print("R^2 Score:", r2)
```

```python
from sklearn.linear_model import Lasso
```

```python
lasso_model=Lasso(alpha=0.1)
lasso_model.fit(x_train,y_train)

y_pred_l1=lasso_model.predict(x_test)
```

```python
mae_L1 = mean_absolute_error(y_test, y_pred_l1)
mse_L1 = mean_squared_error(y_test, y_pred_l1)
rmse_L1 = mean_squared_error(y_test, y_pred_l1, squared=False)  # Calculate RMSE from MSE
r2_L1 = r2_score(y_test, y_pred_l1)

# Print the metrics
print("Mean Absolute Error L1:", mae_L1)
print("Mean Squared Error L1:", mse_L1)
print("Root Mean Squared Error L1:", rmse_L1)
print("R^2 Score L1:", r2_L1)
```

```python
from sklearn.linear_model import Ridge
rigde_model=Ridge(alpha=0.1)
rigde_model.fit(x_train,y_train)

y_pred_l2=rigde_model.predict(x_test)

mae_L2 = mean_absolute_error(y_test, y_pred_l2)
mse_L2 = mean_squared_error(y_test, y_pred_l2)
rmse_L2 = mean_squared_error(y_test, y_pred_l2, squared=False)  # Calculate RMSE from MSE
r2_L2 = r2_score(y_test, y_pred_l2)

# Print the metrics
print("Mean Absolute Error L2:", mae_L2)
print("Mean Squared Error L2:", mse_L2)
print("Root Mean Squared Error L2:", rmse_L2)
print("R^2 Score L2:", r2_L2)
```

```python
from sklearn.linear_model import ElasticNet

EN_model=ElasticNet(alpha=0.2)

EN_model.fit(x_train,y_train)

y_pred_en=EN_model.predict(x_test)

mae_EN = mean_absolute_error(y_test, y_pred_en)
mse_EN = mean_squared_error(y_test, y_pred_en)
rmse_EN = mean_squared_error(y_test, y_pred_en, squared=False)  # Calculate RMSE from MSE
r2_EN = r2_score(y_test, y_pred_en)

# Print the metrics
print("Mean Absolute Error EN:", mae_EN)
print("Mean Squared Error EN:", mse_EN)
print("Root Mean Squared Error L2:", rmse_EN)
print("R^2 Score EN:", r2_EN)
```
